```python
data.head(7)
```

```python
c1=data['V1'].values
c2=data['V2'].values
```

```python
import numpy as np
```

```python
from sklearn.cluster import KMeans     #
```

```python
range_cluster=[1,2,3,5,7,10,15]  #different values of k used to find best fit point in elbow curve
```

```python
sse={}
```

```python
for k in range_cluster:
    kmeans=KMeans(n_clusters=k)
    kmeans_label=kmeans.fit(data)
    sse[k]=kmeans_label.inertia_
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel("number of clusters")
plt.ylabel("inertia")
plt.show()
```

```python
plt.scatter(c1,c2,color="black")
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans_label=kmeans.fit_predict(data)
```

```python
plt.scatter(c1,c2,c=kmeans_label)
```
